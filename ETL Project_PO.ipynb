```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
agents_df = pd.read_csv('Resources/agents.csv')
agencies_df = pd.read_csv('Resources/agencies.csv',names=["agency","agencyURL","NULL","NULL2"])
artists_df = pd.read_csv('Resources/artists.csv')
bands_df = pd.read_csv('Resources/bandsintown.csv', thousands=',')
```

```python
agent_nameFrame = agents_df["First"].str.split(expand=True, n=1)

agents_df["firstName"] = agent_nameFrame[0]
agents_df["lastName"] = agent_nameFrame[1]
agents_df.drop(columns=['First'],inplace=True)
```

```python
agents_df.reset_index(inplace=True)
agents_df = agents_df.rename(columns = {'index':'agentid'})


agents_df = pd.DataFrame({"agentid":agents_df["agentid"],
                         "firstname":agents_df['firstName'],
                         "lastname":agents_df['lastName'],
                         "title":agents_df['Title'],
                         "phone":agents_df["Phone Number"],
                         "email":agents_df["Email"],
                         "address1":agents_df['Street Address'],
                         "address2":agents_df['Street Address 2'],
                         "city":agents_df['City'],
                         "state":agents_df["State"],
                         "zip":agents_df["Zip"],
                         "agency":agents_df["Agency"]})

agents_df.drop_duplicates("agentid", inplace=True)
agents_df.set_index("agentid", inplace=True)
```

```python
art_count = artists_df.groupby(by=['agency']).count()
art_count = art_count['artist'].sort_values(ascending=False)
art_count = art_count.to_frame()
art_count

html_artists = art_count.to_html() 
  
# # write html to file 
text_file = open("artist_count.html", "w") 
text_file.write(html_artists) 
text_file.close() 
```

```python
artists_df['attendance'] = artists_df['attendance'].replace(',', '')
artists_df['averagegross'] = artists_df['averagegross'].str.replace(',', '').str.replace('$', '')

# artists_df['attendance'].dtype

artists_df['averagegross'].dropna(inplace=True)
artists_df['averagegross']

artists_df['attendance'].dropna(inplace=True)
artists_df['attendance']

artist_clean = pd.DataFrame({"Artist":artists_df["artist"],"Attendance":artists_df["attendance"],"Avg Gross":artists_df['averagegross'],"Agency":artists_df["agency"]})

artist_clean.dropna(inplace=True)
agency_attendance = artist_clean.groupby(by=['Agency']).sum().sort_values(by=['Attendance'],ascending=False)

html = agency_attendance.to_html() 
  
# write html to file 
text_file = open("agency_attendance.html", "w") 
text_file.write(html) 
text_file.close() 
```

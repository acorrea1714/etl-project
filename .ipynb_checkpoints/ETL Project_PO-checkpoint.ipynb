```python
import pandas as pd
from sqlalchemy import create_engine
#import psycopg2
```

```python
agents_df = pd.read_csv('Resources/agents.csv')
agencies_df = pd.read_csv('Resources/agencies.csv',names=["agency","agencyURL","NULL","NULL2"])
artists_df = pd.read_csv('Resources/artists.csv')
bands_df = pd.read_csv('Resources/bandsintown.csv', thousands=',')
```

```python
agents_df
```

```python
agent_nameFrame = agents_df["First"].str.split(expand=True, n=1)

agents_df["firstName"] = agent_nameFrame[0]
agents_df["lastName"] = agent_nameFrame[1]
agents_df.drop(columns=['First'],inplace=True)
```

```python
agents_df.reset_index(inplace=True)
agents_df = agents_df.rename(columns = {'index':'agentid'})


agents_df = pd.DataFrame({"agentid":agents_df["agentid"],
                         "firstname":agents_df['firstName'],
                         "lastname":agents_df['lastName'],
                         "title":agents_df['Title'],
                         "phone":agents_df["Phone Number"],
                         "email":agents_df["Email"],
                         "address1":agents_df['Street Address'],
                         "address2":agents_df['Street Address 2'],
                         "city":agents_df['City'],
                         "state":agents_df["State"],
                         "zip":agents_df["Zip"],
                         "agency":agents_df["Agency"]})

agents_df.drop_duplicates("agentid", inplace=True)
agents_df.set_index("agentid", inplace=True)
```

```python
agents_df
```

```python
agencies_df.drop(columns=["NULL","NULL2"],inplace=True)

```

```python
agencies_df.reset_index(inplace=True)
agencies_df = agencies_df.rename(columns = {'index':'agencyid', 'agencyURL':'agencyurl'})

agencies_df.set_index("agencyid", inplace=True)

agencies_df
```

```python
artists_df.drop(columns=["Location"],inplace=True)
```

```python

artists_df.reset_index(inplace=True)
artists_df = artists_df.rename(columns = {'index':'artistid'})
```

```python
artists_df.rename(columns = {'Artist':'artist',"Avg Tickets":"attendance","Gross Sales":"averagegross","Territory":"territory","Agency":"agency"},inplace=True)
artists_df.set_index("artistid", inplace=True)

artists_df
```

```python
bands_df.drop(columns=["Attendance","Gross"],inplace=True)
```

```python
bands_df.rename(columns={"Artist":"artist","Bands in Town":"bandsintown"},inplace=True)
```

```python
agents_df.to_sql(name='agent', con=engine, if_exists='append', index=True)
agencies_df.to_sql(name='agencies', con=engine, if_exists='append', index=True)
artists_df.to_sql(name='artists', con=engine, if_exists='append', index=True)
bands_df.to_sql(name='bands', con=engine, if_exists='append', index=True)
```
